```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import pandas as pd

def gaussian(x, amplitude, mean, stddev):
    return amplitude * np.exp(-((x - mean) ** 2) / (2 * stddev ** 2))

def load_data():
    # Placeholder function to load your data
    # Replace this with the actual data loading code
    x = np.linspace(-5, 5, 100)
    y = gaussian(x, 1, 0, 1) + 0.1 * np.random.normal(size=x.size)
    return x, y

def fit_gaussian(x, y):
    # Initial guesses for fitting parameters
    initial_guesses = [1, np.mean(x), np.std(x)]
    params, _ = curve_fit(gaussian, x, y, p0=initial_guesses)
    return params

def plot_results(x, y, params):
    plt.figure(figsize=(8, 4))
    plt.scatter(x, y, label='Data')
    plt.plot(x, gaussian(x, *params), label='Fitted Gaussian', color='red')
    plt.title('Gaussian Fit')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.legend()
    plt.show()

def calculate_fwhm(stddev):
    return 2 * np.sqrt(2 * np.log(2)) * stddev

df = pd.read_csv("I_CdSe_S3.txt", sep = '\t', decimal= ',' , skiprows = 19, skipfooter=47, header = None)

start = 500
end = 600
filtered_df = df[(df[0] >= start) & (df[0] <= end)]
x = filtered_df[0]
y = filtered_df[1]

plt.plot(x,y)

params = fit_gaussian(x, y)
amplitude, mean, stddev = params
fwhm = calculate_fwhm(stddev)

def calculate_d(x):
    return (59.60816
            - 0.547364 * x
            + 1.8873 * (10 ** -3) * (x ** 2)
            - 2.85743 * (10 ** -6) * (x ** 3)
            + 1.62974 * (10 ** -9) * (x ** 4))

print(f"Maximum of the peak: {amplitude}")
print(f"Mean (Center of Peak): {mean}")
print(f"Particle Diameter: {calculate_d(mean)}")
print(f"Standard Deviation: {stddev}")
print(f"FWHM: {fwhm:.2f}")

plot_results(x, y, params)
```
