```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
import os
import sys
sys.path.append('/home/ABTLUS/victor.secco/data_processing/')
from mypackages.plot_style import set_plot_style
set_plot_style()

Pt4f = pd.read_csv("/home/ABTLUS/victor.secco/data_processing/XPS/Pt4f.csv", sep =";", skiprows=[1], header=0)
Se3d = pd.read_csv("/home/ABTLUS/victor.secco/data_processing/XPS/Se3d.csv", sep =";", skiprows=[1], header=0)


Pt4f["Binding Energy (E)"] = Pt4f["Binding Energy (E)"]/10
Se3d["Binding Energy (E)"] = Se3d["Binding Energy (E)"]/10
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.patches import Patch
from matplotlib.ticker import FuncFormatter
import matplotlib.patheffects as PathEffects

# Example DataFrame (replace this with your actual data)
# Assuming Pt4f is a pandas DataFrame with the necessary columns


# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data
scatter = ax.scatter(Pt4f["Binding Energy (E)"].values, Pt4f["Unnamed: 1"].values, label='XPS Data', color='blue', lw=5)
scatter.set_path_effects([PathEffects.withStroke(linewidth=5, foreground='k')])

ax.plot(Pt4f["Binding Energy (E)"].values, Pt4f["Fitted envelope"].values, label='Se 3d', color='red', lw=3)
ax.plot(Pt4f["Binding Energy (E)"].values, Pt4f["Backgnd."].values, label='Se 3d', color='black', lw=1)

ax.plot(Pt4f["Binding Energy (E)"].values, Pt4f["Fitted Peak Pt4f5 Scan A"].values, label='Pt 4f', color='#F6C385', lw=1)
ax.fill_between(Pt4f["Binding Energy (E)"].values,  Pt4f["Fitted Peak Pt4f5 Scan A"].values, Pt4f["Backgnd."].values,
                 where=(Pt4f["Fitted Peak Pt4f5 Scan A"].values > Pt4f["Backgnd."].values), color='#F6C385', alpha=0.5)

ax.plot(Pt4f["Binding Energy (E)"].values, Pt4f["Fitted Peak Pt4f7 Scan A"].values, label='Pt 4f', color='#F6C385', lw=1)
ax.fill_between(Pt4f["Binding Energy (E)"].values,  Pt4f["Fitted Peak Pt4f7 Scan A"].values, Pt4f["Backgnd."].values,
                 where=(Pt4f["Fitted Peak Pt4f7 Scan A"].values > Pt4f["Backgnd."].values), color='#F6C385', alpha=0.5)

ax.plot(Pt4f["Binding Energy (E)"].values, Pt4f["Fitted Peak Pt4f5 Scan B"].values, label='Pt 4f', color='#42883D', lw=1)
ax.fill_between(Pt4f["Binding Energy (E)"].values,  Pt4f["Fitted Peak Pt4f5 Scan B"].values, Pt4f["Backgnd."].values,
                 where=(Pt4f["Fitted Peak Pt4f5 Scan B"].values > Pt4f["Backgnd."].values), color='#42883D', alpha=0.5)

ax.plot(Pt4f["Binding Energy (E)"].values, Pt4f["Fitted Peak Pt4f7 Scan B"].values, label='Pt 4f', color='#42883D', lw=1)
ax.fill_between(Pt4f["Binding Energy (E)"].values,  Pt4f["Fitted Peak Pt4f7 Scan B"].values, Pt4f["Backgnd."].values,
                 where=(Pt4f["Fitted Peak Pt4f7 Scan B"].values > Pt4f["Backgnd."].values), color='#42883D', alpha=0.5)



# Set the labels and title
ax.set_xlabel('Binding Energy (keV)')
ax.set_ylabel('Intensity (counts)')

plt.xlim(85,64)

def thousands_formatter(x, pos):
    return f'{x*1e-3:.0f}k'

ax.yaxis.set_major_formatter(FuncFormatter(thousands_formatter))

# Customize the legend
legend_handles = [
    plt.Line2D([0], [0], color='blue', marker='o', markersize=8, linestyle='None', label='XPS Data'),
    Patch(facecolor='#FFBE80', edgecolor='#F6C385', label=r'Pt 4$f$ ($\rm{PtSe_x}$)'),
    Patch(facecolor='#7CC87D', edgecolor='#42883D', label=r'Pt 4$f$ ($\rm{PtSe_2}$)'),
    Patch(facecolor='white', edgecolor='black', label='Shirley Bck'),
    plt.Line2D([0], [0], color='red', linestyle='-', label='Fit')
]

ax.legend(handles=legend_handles, frameon=False)

plt.savefig("/home/ABTLUS/victor.secco/data_processing/XPS/Pt4f.png", dpi=500)
# Display the plot
plt.show()
```
