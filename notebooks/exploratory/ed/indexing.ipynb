```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np


# Load and display a .tif image
def display_tif_image(file_path):
    # Open the .tif image
    with Image.open(file_path) as img:
        # Convert image to numpy array for manipulation if needed
        image_array = np.array(img)
        
        # Display basic info about the image
        print(f"Image size: {img.size}")
        print(f"Image mode: {img.mode}")

        # Plot the image
        plt.figure(figsize=(8, 8))
        plt.imshow(image_array, cmap='gray')  # Change colormap as needed
        #plt.colorbar()  # Shows intensity scale
        plt.title(f"Displaying {file_path}")
        plt.axis('off')  # Hide axes for a cleaner view
        plt.show()

# Example usage
display_tif_image("Ref12_reference.tif")
```
