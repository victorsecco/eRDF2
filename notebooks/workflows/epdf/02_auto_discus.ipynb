```python
import subprocess
import pandas as pd
import numpy as np
import os
```

```python
input_path =  "/home/ABTLUS/victor.secco/data_processing/data/processed/epdf/In2O3/iq_In2O3_S4.csv"
df = pd.read_csv(input_path)
output_path = "//home/ABTLUS/victor.secco/data_processing/EXP2PDF/DATA/In2O3"

for i in df.columns:
    iq = df[i]
    q = np.linspace(0, len(iq)*0.00743649587727647, len(iq))
    roi = pd.DataFrame(np.array([q,iq]).T)
    roi.to_csv(os.path.join(output_path,("In2O3_S4_"+i+".csv")), index = None, sep = "\t")
```

```python
file_names = [x for x in os.listdir(output_path)]
file_names.sort()
```

```python
# Open the discus_suite process
process = subprocess.Popen(['wsl', 'discus_suite'],
                            stdin=subprocess.PIPE,
                            stdout=subprocess.PIPE,
                            stderr=subprocess.PIPE, 
                            text=True)

# Send initial setup commands
process.stdin.write('cd ..\n')
process.stdin.write('cd ..\n')
process.stdin.write('cd ..\n')
process.stdin.write('cd EXP2PDF/DATA/In2O3\n')

# Loop through the files and process them one by one
for file in file_names:
    process.stdin.write(f'@prepare.mac {file[:-4]}\n')  # Send the macro command
    process.stdin.flush()  # Flush the input to ensure the command is sent

    # Wait for the process to handle the command by reading the output
    output = process.stdout.read(1024)  # Read partial output to check progress (adjust size if needed)
    print(output)  # Optionally print or log the output
    
    # Optionally, check for specific output or completion messages
    # Here, we assume that the software prints some form of prompt or output when finished.

# Close the process once all files are processed
process.stdin.close()
process.wait()

# Capture final output (if needed)
stdout, stderr = process.communicate()
print(stdout)
print(stderr)
```

```python
import os
import shutil

# Define the paths
source_folder = output_path
destination_folder = os.path.join(output_path, "iqobs")

# Ensure the destination folder exists
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# List and sort all .iqobs files in the source folder
file_names = [x for x in os.listdir(source_folder) if x.endswith('.iqobs')]
file_names.sort()

# Move each file to the destination folder
for file_name in file_names:
    source_path = os.path.join(source_folder, file_name)
    destination_path = os.path.join(destination_folder, file_name)
    
    # Move the file
    shutil.move(source_path, destination_path)
    print(f"Moved {file_name} to {destination_folder}")
```
